```python
import kagglehub

path = kagglehub.dataset_download("mirichoi0218/insurance")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

os.listdir(path)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(os.path.join(path, "insurance.csv"), encoding="latin1")
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True,dtype=float)
```

```python
df.head(3)
```

```python
y=df[['charges']]
x=df.drop('charges',axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[19,26,0,1,1,0,0,1]])
```

```python
#hata şablonu hazırlıyoruz
df_hata=pd.DataFrame()
```

```python
df_hata['y']=y
```

```python
y_tahmin=model.predict(x)
```

```python
df_hata['tahmin']=y_tahmin
```

```python
df_hata
```

```python
df_hata['error']=y-y_tahmin
df_hata.head(10)
```

#MSE hata metriği

```python
df_hata['squared_error']=df_hata['error']**2
```

# MAE Abstract error

```python
#her zaman pozitif değer alması için yapıyoruz
df_hata['abs_error']=np.abs(df_hata['error'])
```

#MAPE

```python
df_hata['percent_error']=np.abs((y-y_tahmin)/y)
```

```python
df_hata.mean()
```

```python
# kısa yoldan yapmak istersek
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
mean_squared_error(y,y_tahmin)
```

```python
mean_absolute_error(y,y_tahmin)
```
