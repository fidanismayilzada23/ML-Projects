```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('/content/Audi_A1_listings.csv')
```

```python
df.head(3)
```

```python
df=df.drop(columns=['index','href','PPY','MileageRank','PriceRank','PPYRank','Score'])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahib","fiyat"]
df
```

```python
df['motor']=df['motor'].str.replace("L","")
```

```python
df['motor']=pd.to_numeric(df['motor'])
```

```python
df.head(3)
```

```python
#ikili objectleri true false yapıyor
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True,dtype=float)
```

```python
df.head(3)
```

```python
y=df[['fiyat']]
x=df.drop("fiyat",axis=1)
```

```python
#datanı hisselere bölerek test ediyoruz ki tam öğrenmesin ve bize daha gerçeğe uygun öğrenme sunsun
#random state her defasında yeni bir rakam getirmemesi için ve farklı öğrenme sunmaması için eklenmesi gerekmektedir
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=22)
```

```python
#Bunu yaparak datamızı daha iyi hale getiriyoruz
lm=LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
